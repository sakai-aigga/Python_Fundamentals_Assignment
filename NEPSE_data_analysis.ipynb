```python
#analyze the index change trend for the last 7 days
#analyze the index change trend for the complete data
#print the number of days when the NEPSE index was greater than 2490 among the last 90 days
#calculate the average NEPSE index value for the last 90 days
#identify the maximum and minimum NEPSE index values within the last 90 days


import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data=pd.read_csv("nepse_last_90_days.csv")


last=data.tail(7)
idx=last["NEPSE Index"]
date=last["Date"]

#analyze index change for last 7 days
plt.plot(date, idx, marker="o")
plt.plot(date, idx, linestyle="dotted")
plt.xlabel('Date')
plt.ylabel('NEPSE Index')
print(idx)
```

```python
#analyze index change trend for the complete date
idx=data["NEPSE Index"]
date=data["Date"]
plt.plot(date, idx, marker="o")
plt.plot(date, idx, linestyle="dotted")
plt.xlabel('Date')
plt.ylabel('NEPSE Index')
print(idx)
```

```python
#print the number of days when the NEPSE index was greater than 2490 among the last 90 days
great_idx=data[data["NEPSE Index"]>2490]
print(great_idx.count())
```

```python
#calculate the average NEPSE index value for the last 90 days
print(data["NEPSE Index"].mean())
```

```python
#identify the maximum and minimum NEPSE index values within the last 90 days
print(data["NEPSE Index"].max())
print(data["NEPSE Index"].min())
```
